```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
tokenizer = AutoTokenizer.from_pretrained("google/pegasus-xsum")
model = AutoModelForSeq2SeqLM.from_pretrained("google/pegasus-xsum")
```

```python
def summerization(text):
  tokens = tokenizer(text, truncation=True, padding="longest", return_tensors="pt")
  summary = model.generate(**tokens)
  return tokenizer.decode(summary[0])
```

```python
!pip install gradio
```

```python
import gradio as gr
desc = "The application will return text in the summarized form"
interface = gr.Interface(fn=summerization, 
                         inputs=gr.inputs.Textbox(lines=2, placeholder='Set the text to summarize'),  outputs='text',
                         description=desc).launch(share=True)
```
